Diseñar un programa en el que entrado un numero a retorne una variable booleana True or false si cumple que es:

**Números defectivo :** la suma de los divisores propios es menor que el número.
La rutina se deberá llamar números_defectivos

**Números abundantes:** la suma de los divisores es mayor que el número.
La rutina se deberá llamar números_abundantes

**Números semiperfectos** la suma de todos o algunos de los divisores propios es igual al número.
La rutina se debera llamar numeros_semiperfectos

**Números perfectos** la suma de todos sus divisores propios, excepto el mismo numero, es igual al número.
La rutina se deberá llamar numeros_perfectos

**Números primos** el número es divisible unicamente por sí mismo y por 1.
La rutina se deberá llamar numeros_primos

**Problema no obligatorio**
Retornar los 30 primeros números de cada clase

```python
def divisores_propios(n):
    return [i for i in range(1, n) if n % i == 0]

def numeros_defectivos(n):
    return sum(divisores_propios(n)) < n

def numeros_abundantes(n):
    return sum(divisores_propios(n)) > n

def numeros_semiperfectos(n):
    divisores = divisores_propios(n)

    def es_subsuma(target, index):
        if target == 0:
            return True
        if target < 0 or index >= len(divisores):
            return False
        # Incluir o no incluir el divisor actual
        return es_subsuma(target - divisores[index], index + 1) or es_subsuma(target, index + 1)

    return es_subsuma(n, 0)

def numeros_perfectos(n):
    return sum(divisores_propios(n)) == n

def numeros_primos(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

def primeros_30_de(clasificador, max_n=100000): #pongo el n max para que no se demore en calcular todos
    resultado = []
    n = 1
    while len(resultado) < 30 and n <= max_n:
        if clasificador(n):
            resultado.append(n)
        n += 1
    return resultado

if __name__ == "__main__":
    print("Primeros 30 números defectivos:", primeros_30_de(numeros_defectivos))
    print("Primeros 30 números abundantes:", primeros_30_de(numeros_abundantes))
    print("Primeros 30 números semiperfectos:", primeros_30_de(numeros_semiperfectos, max_n=10000)[:30])
    print("Primeros 4 números perfectos:", primeros_30_de(numeros_perfectos, max_n=100000)[:4]) # se ponen los primeros 4 ya que mas valores serian muy grandes
    print("Primeros 30 números primos:", primeros_30_de(numeros_primos))
```

Determine si un número n entero ingresado por el usuario es un palíndromo, (Retorne True en caso afirmativo y False en caso contrario)

```python
def es_palindromo(n):
    # Convertimos el número a cadena para poder invertirlo
    n_str = str(n)
    # Comparamos la cadena con su reverso
    return n_str == n_str[::-1]

# Ejemplo de uso
numero = int(input("Ingrese un número entero: "))
if es_palindromo(numero):
    print(f"{numero} es un palíndromo.")
else:
    print(f"{numero} no es un palíndromo.")
```

Un problema conocido de mecánica cuántica involucra una partícula de masa m
 que encuentra un escalón de potencial unidimensional, como se muestra en la figura:
 image.png

 La partícula con energía cinética inicial E y vector de onda k1=√(2mE)/ℏ entra desde la izquierda y encuentra un salto abrupto en la energía potencial de altura V en la posición x=0. Resolviendo la ecuación de Schrödinger, se puede demostrar que cuando E>V
 la partícula puede:
 Pasar el escalón, en cuyo caso tendrá una energía cinética menor E−V
 al otro lado y un vector de onda correspondientemente más pequeño k2=2m√(E−V)/ℏ
Ser reflejada, manteniendo toda su energía cinética y vector de onda sin cambios, pero moviéndose en dirección opuesta.
Las probabilidades T y R para transmisión y reflexión están dadas por:

T=(4k1k2)/(k1+k2)^2, R=(k1−k2/k1+k2)^2.
Consideremos una partícula con masa igual a la del electrón m=9.11×10−31kg
 y energía 10eV que encuentra un escalón de potencial de altura 9eV. Escriba un programa en Python para calcular e imprimir las probabilidades de transmisión y reflexión usando las fórmulas anteriores.

```python
import numpy as np

# Constantes físicas
m = 9.11e-31               # masa del electrón en kg
hbar = 1.055e-34           # h-barra en J·s
eV = 1.602e-19             # 1 eV en Joules

# Energías dadas
E_eV = 10                  # energía total en eV
V_eV = 9                   # escalón de potencial en eV

# Conversión a Joules
E = E_eV * eV
V = V_eV * eV

# Cálculo de vectores de onda
k1 = np.sqrt(2 * m * E) / hbar
k2 = np.sqrt(2 * m * (E - V)) / hbar

# Probabilidades de transmisión y reflexión
T = (4 * k1 * k2) / (k1 + k2)**2
R = ((k1 - k2) / (k1 + k2))**2

# Resultados
print(f"Probabilidad de transmisión (T): {T:.4f}")
print(f"Probabilidad de reflexión (R):   {R:.4f}")
print(f"Verificación: T + R = {T + R:.4f}")
```

La órbita en el espacio de un cuerpo alrededor de otro, como un planeta alrededor del Sol, no necesita ser circular. En general toma la forma de una elipse, con el cuerpo a veces más cerca y a veces más lejos. Si se conocen la distancia ℓ1 del acercamiento más próximo que hace un planeta al Sol, llamado perihelio, y su velocidad lineal v1 en el perihelio, entonces cualquier otra propiedad de la órbita puede calcularse a partir de estos dos datos de la siguiente manera. La segunda ley de Kepler nos dice que la distancia ℓ2 y velocidad v2 del planeta en su punto más distante, o afelio, satisfacen ℓ2v2=ℓ1v1. Al mismo tiempo, la energía total, cinética más gravitacional, de un planeta con velocidad v y distancia r del Sol está dada por:

E=1/2mv^2−G(mM)/r, donde m
 es la masa del planeta, M=1.9891×10^30kg
 es la masa del Sol, y G=6.6738×10^−11 es la constante gravitacional de Newton. Dado que la energía debe conservarse, se llega a que v2
 es la raíz más pequeña de la ecuación cuadrática

 v2^2−((2GM)/(v1ℓ1))v2−[v1^2−(2GM)/ℓ1]=0.

 Una vez que se tiene v2, podemos calcular ℓ2 usando la relación ℓ2=(ℓ1v1)/v2.
Dados los valores de v1, ℓ1, y ℓ2, otros parámetros de la órbita vienen dados por fórmulas simples que pueden derivarse de las leyes de Kepler y del hecho de que la órbita es una elipse:

Semieje mayor:a=1/2(ℓ1+ℓ2),
Semieje menor:b=√(ℓ1ℓ2),
Periodo orbital:T=(2πab)/(ℓ1v1),
Excentricidad orbital:e=(ℓ2−ℓ1)/(ℓ2+ℓ1).

Escriba un programa que solicite al usuario ingresar la distancia al Sol y la velocidad en el perihelio, luego calcule e imprima las cantidades ℓ2, v2, T, y e.

Pruebe su programa calculando las propiedades de las órbitas de la Tierra (para la cual ℓ1=1.4710×10^11m y v1=3.0287×10^4ms^−1) y del cometa Halley (ℓ1=8.7830×10^10m y v1=5.4529×10^4ms^−1). Entre otras cosas, debería encontrar que el período orbital de la Tierra es un año y el del cometa Halley es alrededor de 76 años.


```python
import numpy as np

# Constantes físicas
G = 6.6738e-11           # Constante gravitacional (m^3/kg·s^2)
M = 1.9891e30            # Masa del Sol (kg)

def calcular_orbita(ell1, v1, m=1):  # m es opcional, se cancela en todas partes

    # Coeficientes de la ecuación cuadrática para v2
    A = 1
    B = - (2 * G * M) / (v1 * ell1)
    C = - (v1**2 - (2 * G * M) / ell1)

    # Resolver la ecuación cuadrática A*v2^2 + B*v2 + C = 0
    discriminante = B**2 - 4 * A * C
    if discriminante < 0:
        raise ValueError("Discriminante negativo, no hay soluciones reales para v2.")

    v2_1 = (-B + np.sqrt(discriminante)) / (2 * A)
    v2_2 = (-B - np.sqrt(discriminante)) / (2 * A)

    # Tomamos la raíz más pequeña (la físicamente válida)
    v2 = min(v2_1, v2_2)

    # Calcular l2
    ell2 = (ell1 * v1) / v2

    # Semiejes
    a = 0.5 * (ell1 + ell2)
    b = np.sqrt(ell1 * ell2)

    # Periodo orbital T
    T = (2 * np.pi * a * b) / (ell1 * v1)

    # Excentricidad
    e = (ell2 - ell1) / (ell2 + ell1)

    return ell2, v2, T, e


# Función para imprimir resultados con formato
def imprimir_resultados(nombre, ell1, v1):
    ell2, v2, T, e = calcular_orbita(ell1, v1)
    print(f"--- Órbita de {nombre} ---")
    print(f"Distancia afelio (ℓ2): {ell2:.2e} m")
    print(f"Velocidad afelio (v2): {v2:.2e} m/s")
    print(f"Período orbital (T):   {T / (60*60*24*365.25):.2f} años")
    print(f"Excentricidad (e):     {e:.4f}")
    print()

# Datos de prueba
ell1_tierra = 1.4710e11    # m
v1_tierra = 3.0287e4       # m/s

ell1_halley = 8.7830e10    # m
v1_halley = 5.4529e4       # m/s

# Calcular para la Tierra y el cometa Halley
imprimir_resultados("la Tierra", ell1_tierra, v1_tierra)
imprimir_resultados("el cometa Halley", ell1_halley, v1_halley)
```

Un proyectil es lanzado desde una altura inicial y0=0
 con una velocidad inicial de 4.0m/s a 45^∘ respecto a la horizontal. Graficar la trayectoria del proyectil en su plano de movimiento (considerando el eje x paralelo a la horizontal y el origen en la posición inicial del proyectil) desde el punto de lanzamiento hasta el punto donde impacta el suelo (asumiendo un terreno plano). La gráfica debe incluir etiquetas en los ejes y un título. Recuerde que las ecuaciones del movimiento parabólico están dadas por:

 Posición horizontal:x(t)=v0xt=v0cosθ⋅t
 Posición vertical:y(t)=y0+v0yt−1/2gt^2=y0+v0sinθ⋅t−1/2gt^2
 Velocidad horizontal:vx(t)=v0x=v0cosθ=constante
 Velocidad vertical:vy(t)=v0y−gt=v0sinθ−gt

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos iniciales
v0 = 4.0                      # Velocidad inicial (m/s)
theta_deg = 45               # Ángulo en grados
theta_rad = np.radians(theta_deg)  # Convertir a radianes
y0 = 0                       # Altura inicial
g = 9.81                     # Gravedad (m/s^2)

# Componentes de la velocidad
v0x = v0 * np.cos(theta_rad)
v0y = v0 * np.sin(theta_rad)

# Tiempo de vuelo: resolver y(t) = 0
# y(t) = y0 + v0y * t - 0.5 * g * t^2
# → 0 = v0y * t - 0.5 * g * t^2 → t(v0y - 0.5 * g * t) = 0
# t = 0 o t = 2*v0y/g
t_final = 2 * v0y / g

# Crear array de tiempos
t = np.linspace(0, t_final, 100)

# Calcular posiciones
x = v0x * t
y = y0 + v0y * t - 0.5 * g * t**2

# Graficar trayectoria
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Trayectoria del proyectil', color='darkblue')
plt.xlabel('Distancia horizontal x (m)')
plt.ylabel('Altura vertical y (m)')
plt.title('Movimiento parabólico de un proyectil (45°)')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)  # Línea del suelo
plt.show()
```

Los polinomios de Legendre son de gran importancia en física, especialmente en la solución de la ecuación de Laplace en coordenadas esféricas. Son fundamentales en la expansión en armónicos esféricos, lo que los hace esenciales en problemas de mecánica cuántica y electromagnetismo. Cada polinomio de Legendre Pn(x)es un polinomio de grado n. Este puede ser expresado usando la Fórmula de Rodrigues:

Pn(x)=(1/(n!2^n))d^n/dx^n(x^2−1)^n,  n∈Z+ +{0}

Utilizando la fórmula de Rodrigues halle P0(x)y P1(x).
Con la fórmula de recurrencia:

Pj+1(x)=((2j+1)xPj(x)−jPj−1(x))/j+1,

a) elabore un código que permita calcular Pn(x) para n≥2.

b) En una misma gráfica muestre Pn(x) para 1≤n≤5.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Parte 1: Usar fórmula de Rodrigues para P0(x) y P1(x)
x = sp.symbols('x')

def legendre_rodrigues(n):
    expr = (x**2 - 1)**n
    deriv = sp.diff(expr, x, n)
    Pn = (1 / (2**n * sp.factorial(n))) * deriv
    return sp.simplify(Pn)

P0 = legendre_rodrigues(0)
P1 = legendre_rodrigues(1)

print("P0(x) =", P0)
print("P1(x) =", P1)

# Parte 2a: Calcular Pn(x) para n >= 2 usando recurrencia
def legendre_recurrence(n, x_vals):
    P = [np.ones_like(x_vals), x_vals]  # P0(x), P1(x)
    for j in range(1, n):
        Pj = ((2*j + 1) * x_vals * P[j] - j * P[j-1]) / (j + 1)
        P.append(Pj)
    return P

# Parte 2b: Graficar Pn(x) para n=1 a 5
x_vals = np.linspace(-1, 1, 400)
P_vals = legendre_recurrence(5, x_vals)

plt.figure(figsize=(10, 6))
for n in range(1, 6):
    plt.plot(x_vals, P_vals[n], label=f'P{n}(x)')

plt.title('Polinomios de Legendre $P_n(x)$ para $1 \leq n \leq 5$')
plt.xlabel('x')
plt.ylabel('$P_n(x)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
