**Problema no obligatorio**

````python
def divisores_propios(n):
  '''
  Esta funcion retorna la listas de los divisores de n, salvo el mismo.
  Se recorren todos los numeros del 1 hasta n-1 y se agregan a lista los divisores de n
  '''
  return [x for x in range(1,n) if n % x == 0]



def números_defectivo(a):
  '''
  Se suman los elementos de la lista de divisores y se evalua si cumple con ser un número defectivo
  '''

  divisores = divisores_propios(a)
  b = sum(divisores)
  if b < a:
    return True
  else:
    return False


def números_abundante(a):
  divisores = divisores_propios(a)
  b = sum(divisores)
  if b > a:
    return True
  else:
    return False


def números_perfectos(a):
  divisores = divisores_propios(a)
  b = sum(divisores)
  if b==a:
    return True
  else:
    return False


def números_primos(a):
  '''
  Como un número primo solo es divisible por el mismo y 1, si a es primo entonces en la lista divisores solo esta el 1
  '''

  divisores = divisores_propios(a)
  b = sum(divisores)
  if b == 1:
    return True
  else:
    return False
````

````python
def numeros_semiperfectos(a):
  divisores = divisores_propios(a)

  # Función interna para probar todas las posibles sumas entre divisores propios
  def combinaciones(idx , suma):
    '''
    Es una función recursiva, llamandose a si misma para pobar todas las posibles combinaciones de suma
    Va sumando cada elemnto de la lista, si llega que suma>a entonces no toma el ultimo elemento sumado,
    en cambio pasa al siguiente numero de la lista.
    Si llega al final de la lista y la suma es diferente de a, retorna un falso
    '''

    # Evaluamos el valor de la suma con el número a
    if suma == a:
      return True
    # Si se usaron todos los elementos de la lista o la suma es mayor, ya no se logro llegar al número a
    elif idx == len(divisores) or suma > a:
      return False
    '''
    Aqui se vuelve a llamar la funcion, retornamos el valor de verdad de la afirmación, por lo cual,
    si se intenta con el divisor del idx i+1 en la suma y resulta en suma=a
    o encambio, si se ignora ese valor y da que suma=a, retorna true,
    si ninguna de las dos se cumple por que suma>a o se acabo la lista, retorna falso
    '''
    return (combinaciones(idx + 1, suma + divisores[idx])
            or
            combinaciones(idx + 1, suma))

  # Se retorna el valor booleano dado por "combinaciones", iniciendo la funcion con idx=0 y suma=0
  return combinaciones(0 , 0)
````

Un problema conocido de mecánica cuántica involucra una partícula de masa $m$ que encuentra un escalón de potencial unidimensional, como se muestra en la figura:

image.png


La partícula con energía cinética inicial $E$ y vector de onda $k_1 = \sqrt{2mE}/\hbar$ entra desde la izquierda y encuentra un salto abrupto en la energía potencial de altura $V$ en la posición $x = 0$. Resolviendo la ecuación de Schrödinger, se puede demostrar que cuando $E > V$ la partícula puede:
- Pasar el escalón, en cuyo caso tendrá una energía cinética menor $E - V$ al otro lado y un vector de onda correspondientemente más pequeño $k_2 = \sqrt{2m(E - V)}/\hbar$
- Ser reflejada, manteniendo toda su energía cinética y vector de onda sin cambios, pero moviéndose en dirección opuesta.


Las probabilidades $T$ y $R$ para transmisión y reflexión están dadas por:

$$T = \frac{4k_1k_2}{(k_1 + k_2)^2}, \quad R = \left(\frac{k_1 - k_2}{k_1 + k_2}\right)^2.$$

Consideremos una partícula con masa igual a la del electrón $m = 9.11 \times 10^{-31} \, \text{kg}$ y energía $10 \, \text{eV}$ que encuentra un escalón de potencial de altura $9 \, \text{eV}$. Escriba un programa en Python para calcular e imprimir las probabilidades de transmisión y reflexión usando las fórmulas anteriores.


````python
h = 1.0545718e-34  #Constante de Planck en J*s
eV = 1.60218e-19 #1 eV en joules
m = 9.11e-31 #Masa del electrón en kg

#Energía cinetica y potencial en eV
E_eV = 10
V_eV = 9

# Convertimos a joules
E = E_eV * eV
V = V_eV * eV

k1 = mt.sqrt(2*m*E) / h #Calculamos los vectores de onda k1 y k2
k2 = mt.sqrt(2*m*(E-V)) / h

T = (4 * k1 * k2) / ((k1 + k2)**2) #Probabilidades de transmisión y reflexión
R = ((k1 - k2) / (k1 + k2))**2

print(f"Probabilidad de transmisión T = {T:.4f}")
print(f"Probabilidad de reflexión R = {R:.4f}")
print(f"(T + R) = {T + R:.4f}") #Se verifica que efectivamente la suma de estas dos probabilidades sea igual al 100%
````

La órbita en el espacio de un cuerpo alrededor de otro, como un planeta alrededor del Sol, no necesita ser circular. En general toma la forma de una elipse, con el cuerpo a veces más cerca y a veces más lejos. Si se conocen la distancia $\ell_{1}$ del acercamiento más próximo que hace un planeta al Sol, llamado perihelio, y su velocidad lineal $v_{1}$ en el perihelio, entonces cualquier otra propiedad de la órbita puede calcularse a partir de estos dos datos de la siguiente manera.
La segunda ley de Kepler nos dice que la distancia $\ell_{2}$ y velocidad $v_{2}$ del planeta en su punto más distante, o afelio, satisfacen $\ell_{2}v_{2}=\ell_{1}v_{1}$. Al mismo tiempo, la energía total, cinética más gravitacional, de un planeta con velocidad $v$ y distancia $r$ del Sol está dada por
    $$
    E = \tfrac{1}{2}mv^{2} - G\frac{mM}{r},
    $$
    donde $m$ es la masa del planeta, $M=1.9891\times 10^{30}\,\mathrm{kg}$ es la masa del Sol, y $G=6.6738\times 10^{-11}\,\mathrm{m}^{3}\,\mathrm{kg}^{-1}\,\mathrm{s}^{-2}$ es la constante gravitacional de Newton. Dado que la energía debe conservarse, se llega a que $v_{2}$ es la raíz más pequeña de la ecuación cuadrática
    $$
    v_{2}^{2} - \frac{2GM}{v_{1}\ell_{1}}v_{2} - \left[v_{1}^{2} - \frac{2GM}{\ell_{1}}\right] = 0.
    $$
    Una vez que se tiene $v_{2}$, podemos calcular $\ell_{2}$ usando la relación $\ell_{2}=\ell_{1}v_{1}/v_{2}$.

- Dados los valores de $v_{1}$, $\ell_{1}$, y $\ell_{2}$, otros parámetros de la órbita vienen dados por fórmulas simples que pueden derivarse de las leyes de Kepler y del hecho de que la órbita es una elipse:
$$
    \text{Semieje mayor:}  \quad a = \tfrac{1}{2}(\ell_{1}+\ell_{2}), 
$$
$$
    \text{Semieje menor:}  \quad b = \sqrt{\ell_{1}\ell_{2}}, 
$$
$$
    \text{Periodo orbital:}  \quad T = \frac{2\pi ab}{\ell_{1}v_{1}}, 
$$
$$
    \text{Excentricidad orbital:}  \quad e = \frac{\ell_{2}-\ell_{1}}{\ell_{2}+\ell_{1}}.
$$

Escriba un programa que solicite al usuario ingresar la distancia al Sol y la velocidad en el perihelio, luego calcule e imprima las cantidades $\ell_{2}$, $v_{2}$, $T$, y $e$.

- Pruebe su programa calculando las propiedades de las órbitas de la Tierra (para la cual $\ell_{1}=1.4710\times 10^{11}\,\mathrm{m}$ y $v_{1}=3.0287\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$) y del cometa Halley ($\ell_{1}=8.7830\times 10^{10}\,\mathrm{m}$ y $v_{1}=5.4529\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$). Entre otras cosas, debería encontrar que el período orbital de la Tierra es un año y el del cometa Halley es alrededor de 76 años.


````python
# Importo la constante gravitacional y pi para los calculos y defino la masa del sol = M
from scipy.constants import G, pi

M = 1.9891e30

def orbita(l1, v1):
    '''
    Calcula la distancia del afelio l2, la velocidad en el afelio = v2, el periodo orbital = T y la exentricidad = e
    '''

    # Formulas para v2, l2 y e
    v2 = ((2*G*M)/(v1*l1) + (-v1**2 + (2*G*M)/l1))**(1/2) #Sale de despejar la primera formula
    l2 = (l1*v1)/v2
    e = (l2-l1)/(l2+l1)

    # Variables necesarias para calcular el Periodo
    a = (l1 + l2)/2
    b = (l1*l2)**(1/2)

    T = (2*pi*a*b)/(l1*v1)

    return l2, v2, T, e

# Tierra
l1_Tierra = 1.4710e11
v1_Tierra = 3.0287e4

l2_Tierra , v2_Tierra , T_Tierra , e_Tierra = orbita(l1_Tierra , v1_Tierra)

# Cometa Halley
l1_Halley = 8.7830e10
v1_Halley = 5.4529e4

l2_Halley , v2_Halley , T_Halley , e_Halley = orbita(l1_Halley , v1_Halley)

print(f'Propiedades de la orbita de la Tierra:')
print(f'Afelio={l2_Tierra}, velocidad en el afelio={v2_Tierra}, T={T_Tierra}, e={e_Tierra}')
print(f'Propiedades de la orbita del cometa Halley:')
print(f'Afelio={l2_Halley}, velocidad en el afelio={v2_Halley}, T={T_Halley}, e={e_Halley}')
````

Un proyectil es lanzado desde una altura inicial $y_0 = 0$ con una velocidad inicial de $4.0\, \text{m/s}$ a $45^\circ$ respecto a la horizontal. Graficar la trayectoria del proyectil en su plano de movimiento (considerando el eje $x$ paralelo a la horizontal y el origen en la posición inicial del proyectil) desde el punto de lanzamiento hasta el punto donde impacta el suelo (asumiendo un terreno plano). La gráfica debe incluir etiquetas en los ejes y un título. Recuerde que las ecuaciones del movimiento parabólico están dadas por

$$
\text{Posición horizontal:} \quad x(t) = v_{0x} t = v_0 \cos\theta \cdot t 
$$
$$
\text{Posición vertical:} \quad y(t) = y_0 + v_{0y} t - \frac{1}{2}gt^2 = y_0 + v_0 \sin\theta \cdot t - \frac{1}{2}gt^2 
$$
$$
\text{Velocidad horizontal:} \quad v_x(t) = v_{0x} =v_0 \cos \theta= \text{constante} 
$$
$$
\text{Velocidad vertical:} \quad v_y(t) = v_{0y} - gt = v_0 \sin\theta - gt
$$

````python
def grafica(y0, v0, ang): 
    '''
    Esta función recibe como valores: la altura inicial (y0), la velocidad inicial (v0) 
    y el ángulo de lanzamiento en grados (ang), y grafica la trayectoria del proyectil.
    '''
    import numpy as np
    import matplotlib.pyplot as plt

    ang1 = ang * np.pi / 180  # Conversión de grados a radianes
    g = 9.8
    v0x = v0 * np.cos(ang1)  # Componente horizontal
    v0y = v0 * np.sin(ang1)  # Componente vertical

    # Tiempo total de vuelo (cuando y = 0)
    t_f = (v0y + ((v0y**2) + (2 * g * y0))**0.5) / g
    x_f = v0x * t_f  # Alcance máximo
    ti = np.linspace(0, t_f, 1000)
    yi = y0 + v0y * ti - 0.5 * g * ti**2
    xi = v0x * ti

    fig, ax = plt.subplots()
    ax.plot(xi, yi)
    
    # Añadiendo etiquetas y título
    ax.set_xlabel('Posición horizontal x (m)')
    ax.set_ylabel('Posición vertical y (m)')
    ax.set_title('Trayectoria de un proyectil')
    ax.grid(True)
    plt.tight_layout()

    return plt.show()
````

Los polinomios de Legendre son de gran importancia en física, especialmente en la solución de la ecuación de Laplace en coordenadas esféricas. Son fundamentales en la expansión en armónicos esféricos, lo que los hace esenciales en problemas de mecánica cuántica y electromagnetismo. Cada polinomio de Legendre $P_n(x)$ es un polinomio de grado $n$. Este puede ser expresado usando la Fórmula de Rodrigues:
     
$$
P_n (x)=\frac{1}{n!\ 2^n} \frac{d^n}{dx^n} (x^2-1)^n, \ \ n \in  \mathbb{Z^{+}}+\{0\}
$$
        
- Utilizando la fórmula de Rodrigues halle $P_0(x)$ y $P_1(x)$.
-  Con la fórmula de recurrencia
$$
            P_{j+1}(x)=\frac{(2j+1)x P_j (x)-jP_{j-1}(x)}{j+1},
$$
elabore un código que permita calcular $P_n(x)$ para $n\geq2$.
- En una misma gráfica muestre $P_n(x)$ para $1\leq n\leq5$.


````python
# Declaramos la variable simbólica, esto para imprimir el polinomio sin evaluarlo
x = sp.symbols('x')

def Formula_Rodrigues(n , df_n):
  '''
  Se calcula el coeficeinte 1/n! * 2^n y se multiplica por la derivada.
  Se retorna la multiplicacion simplificada con .simplify
  '''

  coeficiente = 1 / (sp.factorial(n) * 2**n)
  return sp.simplify(coeficiente * df_n)
````

````python
n = 0
f_0 = 1 # Corresponde a (x^2 - 1)^n cuando n=0, y por tanto, su n-esima derivada es ella misma
P_0 = Formula_Rodrigues(n , 1)

print(f'P_{n}(x) =', P_0)
````

````python
n = 1
f_1 = x**2 - 1 # (x^2 - 1)^n con n=1
df_1 = 2 * x   # n-esima derivada de f_1
P_1 = Formula_Rodrigues(n , df_1)

print(f'P_{n}(x) =', P_1)
````

````python
def Legendre(n):
  '''
  Esta funcion utiliza los P_1(x) y P_0(x) encontrados usando la formula de Rodriguez
  como base para encontrar los P_n(x) con n > 2 ó n = 2
  '''

 # Primero evaluemos el valor de n, de modo que si es 1 o 0, retornaremos los polinomios encontrados anteriormente
  if n==0:
    return 1
  elif n==1:
    return x
  else:

    # Comenzamos con los valores iniciales de P_j y P_(j-1)
    Pj= x # P_1(x) = x
    Pj_anterior= 1 # P_0(x)= 1

    for j in range(1 , n):

      # En cuentro P_(j+1) usando la fórmula de recurencia
      Pj_siguiente = ((2*j + 1) * x * Pj - j * Pj_anterior)/(j + 1)

      # Guardo los polinomios P_(j-1) y P_(j) para la siguiente iteracion
      Pj_anterior = Pj
      Pj = Pj_siguiente

    return sp.simplify(Pj)
````

````python
x_vals = np.linspace(-1, 1, 400)

for n in range(1, 6):
    Pn_exprecion = Legendre(n) # Esto nos da una exprecion con valores simbolicos, por tanto, no se puede evaluar
    Pn_funcion = sp.lambdify(x, Pn_exprecion, modules='numpy') # Comvertimos la expresion a una que se puede evaluarla
    y_vals = Pn_funcion(x_vals)
    plt.plot(x_vals, y_vals, label=f'$P_{n}(x)$')


plt.title('Polinomios de Legendre $P_n(x)$ para $1 \\leq n \\leq 5$')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````
