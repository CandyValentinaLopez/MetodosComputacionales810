##2.
El potencial Lennard-Jones es un modelo simple que aproxima la interacción entre un par de moléculas neutras. Este está definido como:

$$
V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r} \right)^{12} - \left( \frac{\sigma}{r} \right)^{6} \right]
$$

Donde:
- $V(r)$ es el potencial en función de la distancia interatómica $r$.
- $\epsilon$ es la profundidad del pozo de potencial, es decir, la energía mínima que describe la interacción entre dos partículas.
- $\sigma$ es la distancia a la cual el potencial es cero, también conocido como el parámetro de distancia.
- $r$ es la distancia entre las dos partículas.

a. Escriba una función para encontrar modelar el potencial LJ.


b. Usando $\sigma = \epsilon = 1$ demuestre que la distancia de equilibrio $r_m$ es aproximadamente $r_m = 2^{1/16}\sigma$. Para esto realice un gráfico del potencial, especificando los ejes. Trunque la serie en $r_c = 2.5\sigma$.


c. Verifique que $r_m = 2^{1/16}\sigma$ se cumple para diferentes valores de $\sigma$, dejando $\epsilon$ constante.




`````python
# a)

def potencial(epsilon,sigma, r):
  v= 4 * epsilon * (((sigma/r)**12) - ((sigma/r)**6))
  return v
`````

`````python
# b)
# Como para este caso sigma=1 entonces el dominio de la funcion va desde 0 hasta 2.5

r = np.linspace(0.1 , 2.5 , 500 )
y = potencial(1 , 1 , r)

plt.ylim(-1.1 , 1.1)
plt.plot(r , y)

plt.xlabel('Distancia')
plt.ylabel('Potencial')
plt.show()
`````

`````python
# La distancia de equilibrio corresponde al minimo de la funcion

min = np.where(y == y.min())[0][0] # Encuentro el indice donde se enceuntro el minimo del potencial
dist_min= r[min]                   #Se encuentra la distancia minima

print(f'La distancia minima corresponde a r = {dist_min}')
`````

`````python
2**(1/6) # Note que el r encontrado graficamente es igual a 2^(1/6)*sigma, en este caso, sigma = 1
`````

`````python
# c)
# Dejando epsilon=1 vamos a variar los valores de sigma en los siguientes graficos

r1 = np.linspace(0.1 , 5 , 500 )
y1 = potencial(1 , 2 , r)

plt.ylim(-1.1 , 1.1)
plt.plot(r1 , y1)

plt.xlabel('Distancia')
plt.ylabel('Potencial')
plt.title('${\sigma} = 2$')
plt.show()

min1 = np.where(y1 == y1.min())[0][0]
dist_min1= r[min1]

print(f'La distancia minima corresponde a r = {dist_min1}')
`````

`````python
r2 = np.linspace(0.1 , 7.5 , 500 )
y2 = potencial(1 , 1/2 , r)

plt.ylim(-1.1 , 1.1)
plt.plot(r2 , y2)

plt.xlabel('Distancia')
plt.ylabel('Potencial')
plt.title('${\sigma} = 1/2$')
plt.show()

min2 = np.where(y2 == y2.min())[0][0]
dist_min2= r[min2]

print(f'La distancia minima corresponde a r = {dist_min2}')
`````
