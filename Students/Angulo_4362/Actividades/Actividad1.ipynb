```python
##Problema 1: Números primos entre 2 y 50 con la Criba de Eratóstenes
def criba_eratostenes(limite):
  primos = [True] * (limite + 1)
  primos[0] = primos[1] = False  # 0 y 1 no son primos

  for i in range(2, int(limite**0.5) + 1):
    if primos[i]:
      for j in range(i * i, limite + 1, i):
        primos[j] = False

  numeros_primos = [i for i, es_primo in enumerate(primos) if es_primo]
  return numeros_primos

# Encuentra los números primos entre 2 y 50
primos_hasta_50 = criba_eratostenes(50)
print("Números primos entre 2 y 50:", primos_hasta_50[2:])
```

```python
##Problema 2: Aproximación de π usando la serie de Leibniz
def approx_π(N):
    suma=0
    for n in range(0,N):
        suma+=4*((-1)**(n))/(2*n+1)
    return suma
print("la aproximacion para n = 10 es:", approx_π(10))
print("la aproximacion para n = 100 es:",approx_π(100))
print("la aproximacion para n = 1000 es:",approx_π(1000))
print("la aproximacion para n = 10000 es:",approx_π(10000))
```

```python
##Problema 3: Números amigos en un rango dado
def suma_divisores_propios(n):
    """Calcula la suma de los divisores propios de un número (excluyendo al mismo número)"""
    if n == 1:
        return 0
    suma = 1
    for i in range(2, n-1):
        if n % i == 0:
            suma += i
    return suma
def encontrar_numeros_amigos(limite_inferior, limite_superior):
    """Encuentra todos los pares de números amigos dentro del rango especificado"""
    pares_amigos = set()

    for a in range(limite_inferior, limite_superior + 1):
        b = suma_divisores_propios(a)

        if (limite_inferior <= b <= limite_superior and suma_divisores_propios(b) == a):
            pares_amigos.add((a, b))

    return sorted(pares_amigos)
print("Buscador de números amigos")
limite_inferior = 1
limite_superior=10000
pares = encontrar_numeros_amigos(limite_inferior, limite_superior)

if pares:
    print("\nPares de números amigos encontrados:")
    for par in pares:
        print(f"{par[0]} y {par[1]}")
else:
    print("\nNo se encontraron pares de números amigos en el rango especificado.")


##Para eliminar los numeros duplicados debo agregar una nueva condicion que sea a < b asi evito que se agregue el par (b,a)
def suma_divisores_propios(n):
    """Calcula la suma de los divisores propios de un número (excluyendo al mismo número)"""
    if n == 1:
        return 0
    suma = 1
    for i in range(2, n-1):
        if n % i == 0:
            suma += i
    return suma
def encontrar_numeros_amigos(limite_inferior, limite_superior):
    """Encuentra todos los pares de números amigos dentro del rango especificado"""
    pares_amigos = set()

    for a in range(limite_inferior, limite_superior + 1):
        b = suma_divisores_propios(a)

        if (limite_inferior <= b <= limite_superior and suma_divisores_propios(b) == a < b):
            pares_amigos.add((a, b))
    return sorted(pares_amigos)

limite_inferior = 1
limite_superior=10000
pares = encontrar_numeros_amigos(limite_inferior, limite_superior)

if pares:
    print("\nPares de números amigos encontrados sin duplicados:")
    for par in pares:
        print(f"{par[0]} y {par[1]}")
else:
    print("\nNo se encontraron pares de números amigos en el rango especificado.")
```

```python
##Problema 4: Contador de frecuencias de letras en una palabra
def contar_frecuencia_letras(palabra):
  frecuencias = {}
  for letra in palabra:
    if 'a' <= letra <= 'z':
      frecuencias[letra] = frecuencias.get(letra, 0) + 1
  return frecuencias

palabra = input("Ingresa una palabra: ")

frecuencias = contar_frecuencia_letras(palabra)

print("Frecuencias de letras:")
for letra, frecuencia in frecuencias.items():
  print(f"{letra}: {frecuencia}")
```
