Diseñar un programa en el que entrado  un numero `a`  retorne una variable booleana True or false si cumple que es:

- [Números defectivo](https://es.wikipedia.org/wiki/Número_defectivo) : la suma de los divisores propios es menor que el número.

  La rutina se deberá llamar números_defectivos



- [Números abundantes](https://es.wikipedia.org/wiki/Número_abundante): la suma de los divisores es mayor que el número.

  La rutina se deberá llamar números_abundantes



- [Números semiperfectos](https://es.wikipedia.org/wiki/N%C3%BAmero_semiperfecto) la suma de todos o algunos de los divisores propios es igual al número.

  La rutina se debera llamar numeros_semiperfectos

- [Números perfectos](https://es.wikipedia.org/wiki/N%C3%BAmero_perfecto) la suma de todos sus divisores propios, excepto el mismo numero, es igual al número.

   La rutina se deberá llamar numeros_perfectos


- [Números primos](https://es.wikipedia.org/wiki/N%C3%BAmero_primo) el número es divisible unicamente por sí mismo y por 1.

   La rutina se deberá llamar numeros_primos


**Problema no obligatorio**

Retornar los 30 primeros números de cada clase

````python
#Aquí va su código
# Función para obtener los divisores propios de un número (excepto él mismo)
def divisores_propios(n):
    return [i for i in range(1, n) if n % i == 0]

# 1. Números defectivos
def numeros_defectivos(n):
    return sum(divisores_propios(n)) < n

# 2. Números abundantes
def numeros_abundantes(n):
    return sum(divisores_propios(n)) > n

# 3. Números semiperfectos
def numeros_semiperfectos(n):
    from itertools import combinations
    divisores = divisores_propios(n)
    for i in range(1, len(divisores) + 1):
        for combo in combinations(divisores, i):
            if sum(combo) == n:
                return True
    return False

# 4. Números perfectos
def numeros_perfectos(n):
    return sum(divisores_propios(n)) == n

# 5. Números primos
def numeros_primos(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Función para obtener los primeros 30 números de cada clase
def primeros_30(funcion_clasificadora):
    numeros = []
    n = 1
    while len(numeros) < 30:
        if funcion_clasificadora(n):
            numeros.append(n)
        n += 1
    return numeros
````

````python
print("Primeros 30 defectivos:", primeros_30(numeros_defectivos))
````

````python
print("Primeros 30 abundantes:", primeros_30(numeros_abundantes))
````

````python
print("Primeros 30 semiperfectos:", primeros_30(numeros_semiperfectos))
````

````python
print("Primeros 30 perfectos:", primeros_30(numeros_perfectos))
````

````python
print("Primeros 30 primos:", primeros_30(numeros_primos))
````

````python
#Aquí va su código
# 1. Cálculo de los Números de Catalan usando recursión
def catalan(n):
    if n == 0:
        return 1
    else:
        return ((4 * n - 2) / (n + 1)) * catalan(n - 1)

# Calculando C50
c50 = catalan(50)
print(f"C_50 es: {c50}")
````

````python
# 2. Cálculo del máximo común divisor (MCD) usando recursión

def gcd(m, n):
    if n == 0:
        return m
    else:
        return gcd(n, m % n)

# Calculando el MCD de 108 y 192
mcd = gcd(108, 192)
print(f"El máximo común divisor de 108 y 192 es: {mcd}")
````

Un problema conocido de mecánica cuántica involucra una partícula de masa $m$ que encuentra un escalón de potencial unidimensional, como se muestra en la figura:

image.png


La partícula con energía cinética inicial $E$ y vector de onda $k_1 = \sqrt{2mE}/\hbar$ entra desde la izquierda y encuentra un salto abrupto en la energía potencial de altura $V$ en la posición $x = 0$. Resolviendo la ecuación de Schrödinger, se puede demostrar que cuando $E > V$ la partícula puede:
- Pasar el escalón, en cuyo caso tendrá una energía cinética menor $E - V$ al otro lado y un vector de onda correspondientemente más pequeño $k_2 = \sqrt{2m(E - V)}/\hbar$
- Ser reflejada, manteniendo toda su energía cinética y vector de onda sin cambios, pero moviéndose en dirección opuesta.


Las probabilidades $T$ y $R$ para transmisión y reflexión están dadas por:

$$T = \frac{4k_1k_2}{(k_1 + k_2)^2}, \quad R = \left(\frac{k_1 - k_2}{k_1 + k_2}\right)^2.$$

Consideremos una partícula con masa igual a la del electrón $m = 9.11 \times 10^{-31} \, \text{kg}$ y energía $10 \, \text{eV}$ que encuentra un escalón de potencial de altura $9 \, \text{eV}$. Escriba un programa en Python para calcular e imprimir las probabilidades de transmisión y reflexión usando las fórmulas anteriores.


## Problema del escalón de potencial en Mecánica Cuántica

Una partícula de masa $( m )$ se mueve en una dimensión y encuentra un **escalón de potencial** como el que se muestra a continuación:



El potencial viene dado por:

$V(x) =
\begin{cases}
0, & x < 0 \\
V_0, & x \geq 0
\end{cases}$

Queremos analizar el comportamiento de la partícula cuando se enfrenta a este cambio repentino en el potencial, considerando dos casos distintos según su energía $( E )$:

---

### ⚛️ Caso 1: Energía mayor que el potencial $( E > V_0 )$

En este caso, la partícula tiene suficiente energía para **cruzar el escalón**.

La función de onda general es:

- Para $( x < 0 )$:

$\psi_I(x) = Ae^{ik_1x} + Be^{-ik_1x}$

- Para $( x \geq 0 )$:

$\psi_{II}(x) = Ce^{ik_2x}$

donde:

$k_1 = \sqrt{\frac{2mE}{\hbar^2}}, \quad k_2 = \sqrt{\frac{2m(E - V_0)}{\hbar^2}}$

> El primer término en $( \psi_I )$ es la onda incidente, el segundo la onda reflejada, y $( \psi_{II} )$ es la onda transmitida.

**Condiciones de frontera:**

En $( x = 0 )$, se imponen:

$\psi_I(0) = \psi_{II}(0), \quad \psi'_I(0) = \psi'_{II}(0)$

De ahí se obtienen las relaciones entre $( A, B, C )$ y los coeficientes de **reflexión** y **transmisión**:

$R = \left| \frac{B}{A} \right|^2, \quad T = \frac{k_2}{k_1} \left| \frac{C}{A} \right|^2$

---

### ⚛️ Caso 2: Energía menor que el potencial $( E < V_0 )$

La partícula no tiene suficiente energía para cruzar el escalón, pero en Mecánica Cuántica **no se detiene**, sino que **penetra una cierta distancia** en la región prohibida.

- Para $( x < 0 )$:

$\psi_I(x) = Ae^{ikx} + Be^{-ikx}, \quad k = \sqrt{\frac{2mE}{\hbar^2}}$

- Para $( x \geq 0 )$:

$\psi_{II}(x) = Ce^{-\kappa x}, \quad \kappa = \sqrt{\frac{2m(V_0 - E)}{\hbar^2}}$

> El término exponencial decreciente describe la **penetración evanescente** en la barrera.

**Condiciones de frontera** iguales a las del caso anterior permiten calcular los coeficientes.

En este caso:

- El coeficiente de **reflexión** es $( R = 1 )$
- El de **transmisión** $( T = 0 )$

pero **la función de onda no se anula inmediatamente** en $( x \geq 0 )$, lo que demuestra el fenómeno de **tunelamiento cuántico**.

---

### 🧠 Conclusión

El análisis del escalón de potencial muestra que incluso un cambio abrupto en el potencial no detiene totalmente una partícula cuántica. A diferencia de la mecánica clásica, **la función de onda penetra en la región prohibida** aunque la probabilidad de encontrar la partícula allí decrezca exponencialmente.

Esto es una manifestación clara de la naturaleza ondulatoria de la materia y del **principio de incertidumbre** de Heisenberg.

````python
#Aquí va su código
import numpy as np

# Constantes
hbar = 1.0545718e-34  # Constante de Planck reducida (J·s)
m = 9.11e-31          # Masa del electrón (kg)

# Parámetros de entrada (puedes modificarlos o pedir al usuario que los ingrese)
E = 5.0  # Energía de la partícula en eV
V0 = 4.0  # Altura de la barrera en eV
L = 1e-9  # Ancho de la barrera en metros

# Conversión de eV a Joules
eV_to_J = 1.602e-19
E_J = E * eV_to_J
V0_J = V0 * eV_to_J

# Cálculo de T y R dependiendo de si E > V0 o E < V0
if E > V0:
    k1 = np.sqrt(2 * m * E_J) / hbar
    k2 = np.sqrt(2 * m * (E_J - V0_J)) / hbar
    T = (4 * k1 * k2) / (k1 + k2)**2
    R = 1 - T
else:
    k1 = np.sqrt(2 * m * E_J) / hbar
    kappa = np.sqrt(2 * m * (V0_J - E_J)) / hbar
    T = np.exp(-2 * kappa * L)
    R = 1 - T

# Impresión de resultados
print(f"Energía de la partícula: {E} eV")
print(f"Altura de la barrera: {V0} eV")
print(f"Ancho de la barrera: {L} m")
print(f"Probabilidad de transmisión (T): {T:.5f}")
print(f"Probabilidad de reflexión (R): {R:.5f}")
````

La órbita en el espacio de un cuerpo alrededor de otro, como un planeta alrededor del Sol, no necesita ser circular. En general toma la forma de una elipse, con el cuerpo a veces más cerca y a veces más lejos. Si se conocen la distancia $\ell_{1}$ del acercamiento más próximo que hace un planeta al Sol, llamado perihelio, y su velocidad lineal $v_{1}$ en el perihelio, entonces cualquier otra propiedad de la órbita puede calcularse a partir de estos dos datos de la siguiente manera.
La segunda ley de Kepler nos dice que la distancia $\ell_{2}$ y velocidad $v_{2}$ del planeta en su punto más distante, o afelio, satisfacen $\ell_{2}v_{2}=\ell_{1}v_{1}$. Al mismo tiempo, la energía total, cinética más gravitacional, de un planeta con velocidad $v$ y distancia $r$ del Sol está dada por
    $$
    E = \tfrac{1}{2}mv^{2} - G\frac{mM}{r},
    $$
    donde $m$ es la masa del planeta, $M=1.9891\times 10^{30}\,\mathrm{kg}$ es la masa del Sol, y $G=6.6738\times 10^{-11}\,\mathrm{m}^{3}\,\mathrm{kg}^{-1}\,\mathrm{s}^{-2}$ es la constante gravitacional de Newton. Dado que la energía debe conservarse, se llega a que $v_{2}$ es la raíz más pequeña de la ecuación cuadrática
    $$
    v_{2}^{2} - \frac{2GM}{v_{1}\ell_{1}}v_{2} - \left[v_{1}^{2} - \frac{2GM}{\ell_{1}}\right] = 0.
    $$
    Una vez que se tiene $v_{2}$, podemos calcular $\ell_{2}$ usando la relación $\ell_{2}=\ell_{1}v_{1}/v_{2}$.

- Dados los valores de $v_{1}$, $\ell_{1}$, y $\ell_{2}$, otros parámetros de la órbita vienen dados por fórmulas simples que pueden derivarse de las leyes de Kepler y del hecho de que la órbita es una elipse:
$$
    \text{Semieje mayor:}  \quad a = \tfrac{1}{2}(\ell_{1}+\ell_{2}),
$$
$$
    \text{Semieje menor:}  \quad b = \sqrt{\ell_{1}\ell_{2}},
$$
$$
    \text{Periodo orbital:}  \quad T = \frac{2\pi ab}{\ell_{1}v_{1}},
$$
$$
    \text{Excentricidad orbital:}  \quad e = \frac{\ell_{2}-\ell_{1}}{\ell_{2}+\ell_{1}}.
$$

Escriba un programa que solicite al usuario ingresar la distancia al Sol y la velocidad en el perihelio, luego calcule e imprima las cantidades $\ell_{2}$, $v_{2}$, $T$, y $e$.

- Pruebe su programa calculando las propiedades de las órbitas de la Tierra (para la cual $\ell_{1}=1.4710\times 10^{11}\,\mathrm{m}$ y $v_{1}=3.0287\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$) y del cometa Halley ($\ell_{1}=8.7830\times 10^{10}\,\mathrm{m}$ y $v_{1}=5.4529\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$). Entre otras cosas, debería encontrar que el período orbital de la Tierra es un año y el del cometa Halley es alrededor de 76 años.


````python
#Aquí va su código
import math

# Constantes
G = 6.6738e-11  # Constante de gravitación universal (m^3 kg^-1 s^-2)
M = 1.9891e30   # Masa del Sol (kg)

def calcular_orbita(l1, v1):
    # Coeficientes de la ecuación cuadrática
    A = 1
    B = -(2 * G * M) / (v1 * l1)
    C = -(v1**2 - (2 * G * M) / l1)

    # Resolver la ecuación cuadrática
    discriminante = B**2 - 4 * A * C
    if discriminante < 0:
        raise ValueError("No hay solución real para v2.")

    v2_pos = (-B + math.sqrt(discriminante)) / (2 * A)
    v2_neg = (-B - math.sqrt(discriminante)) / (2 * A)

    # Escoger la raíz positiva más pequeña
    v2 = min(filter(lambda x: x > 0, [v2_pos, v2_neg]))

    # Calcular l2
    l2 = l1 * v1 / v2

    # Semieje mayor y menor
    a = 0.5 * (l1 + l2)
    b = math.sqrt(l1 * l2)

    # Periodo orbital
    T = (2 * math.pi * a * b) / (l1 * v1)

    # Excentricidad
    e = (l2 - l1) / (l2 + l1)

    return l2, v2, T, e

# ENTRADA MANUAL

print("Ingrese los datos del objeto (en unidades SI):")
try:
    l1_usuario = float(input("Distancia al Sol en el perihelio (l1) en metros: "))
    v1_usuario = float(input("Velocidad en el perihelio (v1) en m/s: "))

    l2, v2, T, e = calcular_orbita(l1_usuario, v1_usuario)

    print(f"\n=== Resultados del objeto ingresado ===")
    print(f"Distancia en el afelio (l2): {l2:.3e} m")
    print(f"Velocidad en el afelio (v2): {v2:.3e} m/s")
    print(f"Periodo orbital (T): {T:.2f} s ({T/(60*60*24*365.25):.2f} años)")
    print(f"Excentricidad orbital (e): {e:.5f}")
except ValueError as err:
    print(f"Error: {err}")

# CASOS DE PRUEBA: TIERRA Y HALLEY

datos = {
    "Tierra": {
        "l1": 1.4710e11,
        "v1": 3.0287e4
    },
    "Cometa Halley": {
        "l1": 8.7830e10,
        "v1": 5.4529e4
    }
}

for nombre, valores in datos.items():
    l1 = valores["l1"]
    v1 = valores["v1"]
    l2, v2, T, e = calcular_orbita(l1, v1)

    print(f"\n=== {nombre} ===")
    print(f"Distancia en el afelio (l2): {l2:.3e} m")
    print(f"Velocidad en el afelio (v2): {v2:.3e} m/s")
    print(f"Periodo orbital (T): {T:.2f} s ({T/(60*60*24*365.25):.2f} años)")
    print(f"Excentricidad orbital (e): {e:.5f}")
````

Los polinomios de Legendre son de gran importancia en física, especialmente en la solución de la ecuación de Laplace en coordenadas esféricas. Son fundamentales en la expansión en armónicos esféricos, lo que los hace esenciales en problemas de mecánica cuántica y electromagnetismo. Cada polinomio de Legendre $P_n(x)$ es un polinomio de grado $n$. Este puede ser expresado usando la Fórmula de Rodrigues:
     
$$
P_n (x)=\frac{1}{n!\ 2^n} \frac{d^n}{dx^n} (x^2-1)^n, \ \ n \in  \mathbb{Z^{+}}+\{0\}
$$
        
- Utilizando la fórmula de Rodrigues halle $P_0(x)$ y $P_1(x)$.
-  Con la fórmula de recurrencia
$$
            P_{j+1}(x)=\frac{(2j+1)x P_j (x)-jP_{j-1}(x)}{j+1},
$$
elabore un código que permita calcular $P_n(x)$ para $n\geq2$.
- En una misma gráfica muestre $P_n(x)$ para $1\leq n\leq5$.


## Cálculo de los Polinomios de Legendre

La fórmula de **Rodrigues** para los polinomios de Legendre es:

$$
P_n (x) = \frac{1}{n! \, 2^n} \frac{d^n}{dx^n} (x^2 - 1)^n
$$

### 1. Cálculo de $( P_0(x) )$ y $( P_1(x) )$ con la fórmula de Rodrigues

Primero, calculemos $(P_0(x))$ y $(P_1(x))$:

#### Para $(P_0(x))$:
Usando la fórmula de Rodrigues con $( n = 0 )$:

$$
P_0(x) = \frac{1}{0! \, 2^0} \frac{d^0}{dx^0} (x^2 - 1)^0 = 1
$$

Por lo tanto, $( P_0(x) = 1 )$.

#### Para $( P_1(x) )$:
Usando la fórmula de Rodrigues con $( n = 1 )$:

$$
P_1(x) = \frac{1}{1! \, 2^1} \frac{d}{dx} (x^2 - 1)^1 = \frac{1}{2} \cdot 2x = x
$$

Por lo tanto, $( P_1(x) = x )$.

### 2. Cálculo de $( P_n(x) )$ para $( n \geq 2 )$ usando la fórmula de recurrencia

La fórmula de recurrencia que nos dan es:

$$
P_{j+1}(x) = \frac{(2j+1)x P_j(x) - j P_{j-1}(x)}{j+1}
$$

Comenzamos con los valores base:
- $( P_0(x) = 1 )$
- $( P_1(x) = x )$

Ahora, podemos usar la fórmula de recurrencia para calcular $( P_2(x) )$, $( P_3(x) )$, $( P_4(x) )$ y $( P_5(x) )$.

````python
#Aquí va su código
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular P_n(x) usando la fórmula de recurrencia
def legendre_polynomial(n, x):
    if n == 0:
        return np.ones_like(x)  # P_0(x) = 1
    elif n == 1:
        return x  # P_1(x) = x

    # Inicializamos P_0(x) = 1 y P_1(x) = x
    P0 = np.ones_like(x)
    P1 = x

    # Usamos la fórmula de recurrencia para obtener P_n(x)
    for j in range(1, n):
        P_next = ((2 * j + 1) * x * P1 - j * P0) / (j + 1)
        P0 = P1
        P1 = P_next

    return P1

# Definir el rango de x
x = np.linspace(-1, 1, 400)

# Graficar P_n(x) para n=1 a n=5
plt.figure(figsize=(8, 6))

for n in range(1, 6):
    Pn = legendre_polynomial(n, x)
    plt.plot(x, Pn, label=f'P_{n}(x)')

# Añadir etiquetas y título
plt.title('Polinomios de Legendre para n=1 a n=5')
plt.xlabel('x')
plt.ylabel('P_n(x)')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
````
