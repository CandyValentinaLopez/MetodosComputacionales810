Diseñar un programa en el que entrado un numero a retorne una variable booleana True or false si cumple que es:

**Números defectivo** : la suma de los divisores propios es menor que el número.

La rutina se deberá llamar números_defectivos

**Números abundantes**: la suma de los divisores es mayor que el número.

La rutina se deberá llamar números_abundantes

**Números semiperfectos** la suma de todos o algunos de los divisores propios es igual al número.

La rutina se debera llamar numeros_semiperfectos

**Números perfectos** la suma de todos sus divisores propios, excepto el mismo numero, es igual al número.

La rutina se deberá llamar numeros_perfectos

**Números primos** el número es divisible unicamente por sí mismo y por 1.

La rutina se deberá llamar numeros_primos

**Problema no obligatorio**

**Retornar los 30 primeros números de cada clase**

```
def divisores_propios(n):
    return [i for i in range(1, n) if n % i == 0]

def numeros_defectivos(n):
    return sum(divisores_propios(n)) < n

def numeros_abundantes(n):
    return sum(divisores_propios(n)) > n

def numeros_semiperfectos(n):
    from itertools import combinations
    divs = divisores_propios(n)
    for r in range(1, len(divs) + 1):
        for subset in combinations(divs, r):
            if sum(subset) == n:
                return True
    return False

def numeros_perfectos(n):
    return sum(divisores_propios(n)) == n

def numeros_primos(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def primeros_30(funcion, limite=10000):
    resultado = []
    for n in range(1, limite):
        if funcion(n):
            resultado.append(n)
            if len(resultado) == 30:
                break
    return resultado

if __name__ == "__main__":
    print("30 primeros números defectivos:", primeros_30(numeros_defectivos))
    print("30 primeros números abundantes:", primeros_30(numeros_abundantes))
    print("30 primeros números semiperfectos:", primeros_30(numeros_semiperfectos))
    print("Números perfectos encontrados (hasta 10000):", primeros_30(numeros_perfectos, limite=10000))
    print("30 primeros números primos:", primeros_30(numeros_primos))
```

 entra desde la izquierda y encuentra un salto abrupto en la energía potencial de altura V
 en la posición x=0
. Resolviendo la ecuación de Schrödinger, se puede demostrar que cuando E>V
 la partícula puede:

Pasar el escalón, en cuyo caso tendrá una energía cinética menor E−V
 al otro lado y un vector de onda correspondientemente más pequeño k2=2m(E−V)−−−−−−−−−√/ℏ
Ser reflejada, manteniendo toda su energía cinética y vector de onda sin cambios, pero moviéndose en dirección opuesta.
Las probabilidades T
 y R
 para transmisión y reflexión están dadas por:

T=4k1k2(k1+k2)2,R=(k1−k2k1+k2)2.
Consideremos una partícula con masa igual a la del electrón m=9.11×10−31kg
 y energía 10eV
 que encuentra un escalón de potencial de altura 9eV
. Escriba un programa en Python para calcular e imprimir las probabilidades de transmisión y reflexión usando las fórmulas anteriores.

```
import math

m = 9.11e-31
hbar = 1.055e-34
eV_to_J = 1.602e-19

E_eV = 10
V_eV = 9

E = E_eV * eV_to_J
V = V_eV * eV_to_J

k1 = math.sqrt(2 * m * E) / hbar
k2 = math.sqrt(2 * m * (E - V)) / hbar
T = (4 * k1 * k2) / ((k1 + k2) ** 2)
R = ((k1 - k2) / (k1 + k2)) ** 2
print(f"Probabilidad de transmisión T = {T:.4f}")
print(f"Probabilidad de reflexión R = {R:.4f}")
print(f"Verificación: T + R = {T + R:.4f}")
```

La órbita en el espacio de un cuerpo alrededor de otro, como un planeta alrededor del Sol, no necesita ser circular. En general toma la forma de una elipse, con el cuerpo a veces más cerca y a veces más lejos. Si se conocen la distancia ℓ1
 del acercamiento más próximo que hace un planeta al Sol, llamado perihelio, y su velocidad lineal v1

 y velocidad v2
 del planeta en su punto más distante, o afelio, satisfacen ℓ2v2=ℓ1v1
. Al mismo tiempo, la energía total, cinética más gravitacional, de un planeta con velocidad v
 y distancia r
 del Sol está dada por
E=12mv2−GmMr,
donde m
 es la masa del planeta, M=1.9891×1030kg
 es la masa del Sol, y G=6.6738×10−11m3kg−1s−2
 es la constante gravitacional de Newton. Dado que la energía debe conservarse, se llega a que v2
 es la raíz más pequeña de la ecuación cuadrática
v22−2GMv1ℓ1v2−[v21−2GMℓ1]=0.
Una vez que se tiene v2
, podemos calcular ℓ2
 usando la relación ℓ2=ℓ1v1/v2

Dados los valores de v1
, ℓ1
, y ℓ2
, otros parámetros de la órbita vienen dados por fórmulas simples que pueden derivarse de las leyes de Kepler y del hecho de que la órbita es una elipse:
Semieje mayor:a=12(ℓ1+ℓ2),

Periodo orbital:T=2πabℓ1v1,

Escriba un programa que solicite al usuario ingresar la distancia al Sol y la velocidad en el perihelio, luego calcule e imprima las cantidades ℓ2
, v2
, T
, y e
.

Pruebe su programa calculando las propiedades de las órbitas de la Tierra (para la cual ℓ1=1.4710×1011m
 y v1=3.0287×104ms−1
) y del cometa Halley (ℓ1=8.7830×1010m
 y v1=5.4529×104ms−1
). Entre otras cosas, debería encontrar que el período orbital de la Tierra es un año y el del cometa Halley es alrededor de 76 años.

```
import math

G = 6.6738e-11
M = 1.9891e30

def calcular_parametros_orbitales(l1, v1):
    a = 2*G*M/l1 - v1**2
    b = -2 * G * M
    c = (l1 * v1)**2

    D = b*b - 4*a*c
    if D < 0:
        raise ValueError("No hay soluciones reales")

    l2_1 = (-b + math.sqrt(D)) / (2*a)
    l2_2 = (-b - math.sqrt(D)) / (2*a)
    l2 = max(l2_1, l2_2)
    v2 = l1 * v1 / l2
    a_semieje = 0.5 * (l1 + l2)
    b_semieje = math.sqrt(l1 * l2)
    T = 2 * math.pi * math.sqrt(a_semieje**3 / (G * M))
    e = (l2 - l1) / (l2 + l1)

    return l2, v2, a_semieje, b_semieje, T, e

def imprimir_resultados(nombre, l1, v1):
    l2, v2, a_s, b_s, T_s, e = calcular_parametros_orbitales(l1, v1)
    print(f"Órbita de {nombre}:")
    print(f"  l2   = {l2} m")
    print(f"  v2   = {v2} m/s")
    print(f"  a    = {a_s} m")
    print(f"  b    = {b_s} m")
    print(f"  T    = {T_s / (60*60*24*365.25)} años")
    print(f"  e    = {e}")
    print()

if __name__ == "__main__":
    l1_tierra = 1.4710e11
    v1_tierra = 3.0287e4
    imprimir_resultados("la Tierra", l1_tierra, v1_tierra)
    l1_halley = 8.7830e10
    v1_halley = 5.4529e4
    imprimir_resultados("el cometa Halley", l1_halley, v1_halley)
```

 con una velocidad inicial de 4.0m/s
 a 45∘
 respecto a la horizontal. Graficar la trayectoria del proyectil en su plano de movimiento (considerando el eje x
 paralelo a la horizontal y el origen en la posición inicial del proyectil) desde el punto de lanzamiento hasta el punto donde impacta el suelo (asumiendo un terreno plano). La gráfica debe incluir etiquetas en los ejes y un título. Recuerde que las ecuaciones del movimiento parabólico están dadas por

Posición horizontal:x(t)=v0xt=v0cosθ⋅t
Posición vertical:y(t)=y0+v0yt−12gt2=y0+v0sinθ⋅t−12gt2
Velocidad horizontal:vx(t)=v0x=v0cosθ=constante
Velocidad vertical:vy(t)=v0y−gt=v0sinθ−gt

```
import numpy as np
import matplotlib.pyplot as plt

v0 = 4.0
theta_deg = 45
theta_rad = np.radians(theta_deg)
y0 = 0.0
g = 9.81

v0x = v0 * np.cos(theta_rad)
v0y = v0 * np.sin(theta_rad)

t_flight = 2 * v0y / g

t = np.linspace(0, t_flight, num=100)

x = v0x * t
y = y0 + v0y * t - 0.5 * g * t**2

plt.plot(x, y)
plt.title("Trayectoria de un proyectil lanzado 45° a 4.0 m/s")
plt.xlabel("Distancia horizontal x (m)")
plt.ylabel("Altura y (m)")
plt.grid(True)
plt.axhline(0, color='black')
plt.show()
```

Los polinomios de Legendre son de gran importancia en física, especialmente en la solución de la ecuación de Laplace en coordenadas esféricas. Son fundamentales en la expansión en armónicos esféricos, lo que los hace esenciales en problemas de mecánica cuántica y electromagnetismo. Cada polinomio de Legendre Pn(x)
 es un polinomio de grado n
. Este puede ser expresado usando la Fórmula de Rodrigues:

Pn(x)=1n! 2ndndxn(x2−1)n,  n∈Z++{0}
Utilizando la fórmula de Rodrigues halle P0(x)
 y P1(x)
.
Con la fórmula de recurrencia
Pj+1(x)=(2j+1)xPj(x)−jPj−1(x)j+1,
elabore un código que permita calcular Pn(x)
 para n≥2
.

En una misma gráfica muestre Pn(x)
 para 1≤n≤5
.

```
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

def rodrigues(n, x):
    from numpy.polynomial.polynomial import Polynomial
    from numpy import polyder, polyval

    poly = Polynomial.fromroots([-1, 1]) ** n

    for _ in range(n):
        poly = poly.deriv()
    return (1 / factorial(n)) * (2**n) * poly(x)

def P0(x):
    return rodrigues(0, x)

def P1(x):
    return rodrigues(1, x)

def legendre(n, x):
    if n == 0:
        return P0(x)
    elif n == 1:
        return P1(x)
    else:
        P_nm1 = P0(x)
        P_n = P1(x)
        for j in range(1, n):
            P_np1 = ((2*j + 1)*x*P_n - j*P_nm1)/(j + 1)
            P_nm1, P_n = P_n, P_np1
        return P_n

x = np.linspace(-1, 1, 400)

plt.figure(figsize=(8,6))
for n in range(1, 6):
    y = legendre(n, x)
    plt.plot(x, y, label=f'P{n}(x)')

plt.title('Polinomios de Legendre')
plt.xlabel('x')
plt.ylabel('P_n(x)')
plt.grid(True)
plt.legend()
plt.show()
```
