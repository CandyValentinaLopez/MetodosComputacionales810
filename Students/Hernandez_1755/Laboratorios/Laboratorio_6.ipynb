Cuatro bloques de masas diferentes $m_i$ están conectados por cuerdas de masa despreciable. Tres de los bloques se encuentran sobre un plano inclinado con un ángulo $\theta$, donde los coeficientes de fricción entre los bloques y el plano son $\mu_i$. Las ecuaciones de movimiento para el sistema son:

$$
T_1 + m_1a = m_1g(\sin \theta - \mu_1 \cos \theta)
$$
$$
-T_1 + T_2 + m_2a = m_2g(\sin \theta - \mu_2 \cos \theta)
$$
$$
-T_2 + T_3 + m_3a = m_3g(\sin \theta - \mu_3 \cos \theta)
$$
$$
-T_3 + m_4a = -m_4g
$$

donde:
- $T_i$ representan las tensiones en las cuerdas
- $a$ es la aceleración del sistema
- $\theta = 45^\circ$ es el ángulo de inclinación
- $g = 9.81 m/s^2$ es la aceleración gravitatoria


Determine $a$ y $T_i$

**Datos:**

Masas de los bloques:
$$
\mathbf{m} = \begin{bmatrix}
10 & 4 & 5 & 6 \\
\end{bmatrix}^T kg
$$

Coeficientes de fricción:
$$
\mu = \begin{bmatrix}
0.25 & 0.3 & 0.2 \\
\end{bmatrix}^T
$$


```python
#Aquí va su código
import numpy as np
g= 9.81
theta= np.pi/4
m=np.array([10,4,5,6])
u=np.array([0.25,0.3,0.2])
# defino los valores que voy a usar junto con las matrices para definir el sistema Ax = b
matriz= np.array([[1,0,0,m[0]],
                  [-1,1,0,m[1]],
                  [0,-1,1,m[2]],
                  [0,0,-1,m[3]]])
b = np.array([
    m[0]*g*(np.sin(theta) - u[0]*np.cos(theta)),
    m[1]*g*(np.sin(theta) - u[1]*np.cos(theta)),
    m[2]*g*(np.sin(theta) - u[2]*np.cos(theta)),
    -m[3]*g])
x= np.linalg.solve(matriz,b)

print(f'T1 = {x[0]} N, T2 = {x[1]} N, T3 = {x[2]} N, a = {x[3]} m/s**2')
```

Se sabe que un sistema físico es gobernado por las ecuaciones,

$$
\begin{matrix}
x' &=& 2x&-y&+z, \\
y' &=&   &3y&-z, \\
z' &=& 2x&+y&+3z,
\end{matrix}
$$

donde $x',y',z'$ son las derivadas, en forma vectorial esto se puede representar como $\mathbf{r}'=A\mathbf{r}$ que tiene solución analítica general, $\mathbf{r} = e^{At} \mathbf{r}(0)$.

- Realice la diagonalización de $A$.

- Use el hecho de que $f(A) =U f(D) U^{-1}$ para encontrar $x(t),y(t),z(t)$ en el intervalo de tiempo $[0,2]$.

- Grafique $x(t),y(t),z(t)$ (use la condición inicial $\mathbf{r}(0)=(1,1,2))$.

- Compare el resultado a la solución analítica,

$$
2x  = x(0)(e^{2t}(1+e^{2t}-2t)) + y(0) (-2te^{2t}) + z(0)(e^{2t}(-1+e^{2t})),
$$

$$
2y  = x(0)(-e^{2t}(-1+e^{2t}-2t)) + y(0)(2(t+1)e^{2t}) + z(0)(-e^{2t}(-1+e^{2t})),
$$

$$
2z  = x(0)(e^{2t}(-1+e^{2t}+2t)) + y(0)(2te^{2t}) + z(0)(e^{2t}(1+e^{2t})).
$$

```python
#Aquí va su código
import matplotlib.pyplot as plt

# defino la matriz el problema planteado
A = np.array([[2, -1, 1],
              [0, 3, -1],
              [2, 1, 3]], dtype=float)
x0, y0, z0 = 1, 1, 2
r0 = np.array([x0, y0, z0], dtype=float)
t = np.linspace(0, 2, 100)
# usando linalg.eig defino D y U como matrices de valores propios y los vectores propios para realizar los calculos después
D, U = np.linalg.eig(A)
U_inv = np.linalg.inv(U) # invierto la matriz de vectore propios

# hago una función para facilitar en el plot los valores próximos usando un for
def r_t(ti):
    return U @ np.diag(np.exp(D*ti)) @ U_inv @ r0 # usando las formulas de r = eAt r(0) y f(A)= Uf(D)U-1 muestro a rprima igual en cada punto como el producto punto de la matriz de vectores propios por la diagonal con los valores propios por la inversa y al final por el vector r0

r_sol = np.array([r_t(ti) for ti in t]) # las soluciones para cada ti será la función r_t en cada valor ti
x_num, y_num, z_num = r_sol[:,0], r_sol[:,1], r_sol[:,2] # simplemente nombro los valores x,y,z en cada punto de la solucion para asignarlos por separado y ver el plot

# la slución analitica mostrada en el trabajo multiplicada por 0.5 para poder igualar bien todo
x_ana = (x0*(np.exp(2*t)*(1+np.exp(2*t)-2*t)) + y0*(-2*t*np.exp(2*t)) + z0*(np.exp(2*t)*(-1+np.exp(2*t))))*0.5
y_ana = (x0*(-np.exp(2*t)*(-1+np.exp(2*t)-2*t)) + y0*(2*(t+1)*np.exp(2*t)) + z0*(-np.exp(2*t)*(-1+np.exp(2*t))))*0.5
z_ana = (x0*(np.exp(2*t)*(-1+np.exp(2*t)+2*t)) + y0*(2*t*np.exp(2*t)) + z0*(np.exp(2*t)*(1+np.exp(2*t))))*0.5

# ploteo


plt.plot(t, x_num, 'b', label='x(t) numérica')
plt.plot(t, y_num, 'g', label='y(t) numérica')
plt.plot(t, z_num, 'r', label='z(t) numérica')

plt.plot(t, x_ana, 'y--', label='x(t) analítica')
plt.plot(t, y_ana, 'y--', label='y(t) analítica')
plt.plot(t, z_ana, 'y--', label='z(t) analítica')
plt.legend()
plt.grid(True)
plt.show()

print(f'los errores en cada componente son: {x_num-x_ana} en x, {y_num-y_ana} en y, y {z_num-z_ana} en z')
# sus errores son minimos en cada iteración
# las graficas se superponen
```

Un electrón interactúa con un campo magnético uniforme. Para determinar los niveles de energía permitidos del electrón en presencia del campo magnético, es necesario resolver la ecuación:

$$
\hat{H}|\Psi\rangle = E|\Psi\rangle
$$

donde el Hamiltoniano está dado por $H = -{\mu} \cdot {B} = -\gamma {B} \cdot \ {S}$, siendo $\gamma$ la razón giromagnética, $\ {B}$ el campo magnético y $\ {S}$ el espín. Puede demostrarse que el Hamiltoniano se expresa como:

$$
\hat{H} = -\frac{\gamma\hbar}{2}\begin{pmatrix}
B_z & B_x - iB_y \\
B_x + iB_y & -B_z
\end{pmatrix}
$$

El valor experimental de $\gamma$ para el electrón es 2. El orden de magnitud del campo magnético es 1 gauss en dirección z.

- Verifique que $H$ es hermítica.
- Encuentre los niveles de energía permitidos.
- Determine los autovectores y verifique su normalización.



```python
#Aquí va su código
# el enunciado dice que el campo magnetico tiene magnitud 1 en la dirección z asi que B.
B= np.array([0,0,1])
gamma= 2
hbarra= 1
# defino el hamiltoniano tal como se menciona en el problema
h = -gamma*hbarra*0.5*np.array([[1,0],
 [0,-1]]) # como el bx y el by valen 0, el hermitiano queda multiplicando bz que es 1 por el sigma z del espin que es la tercera matriz de pauli y queda así
autovalores,autovectores=np.linalg.eig(h) # los autovalores son los niveles de energía permitidos asi que defino el primer valor de np.linalg.eig como autovalores para los niveles de energía y los autovectores como el segundo valor
print(f'hamiltoniano {h}')
print(f'')
print(f'el valor de verdad del hermetismo es {np.allclose(h,np.transpose(h))}') #
print(f'')
print(f'los niveles de energía permitidos son {autovalores}')
print(f'')
print(f'los autovectores son {autovectores}')
print(f'las normas son {np.linalg.norm(autovectores[:,0])} y {np.linalg.norm(autovectores[:,1])}')
```
